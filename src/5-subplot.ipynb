### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 50)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots (2x2 layout)
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line plot
axes[0, 0].plot(x, y, color='b', linestyle='-', linewidth=2)
axes[0, 0].set_title("Line Plot")

# Second subplot: Bar plot
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")

# Third subplot: Scatter plot
axes[1, 0].scatter(x_scatter, y_scatter, color='r', marker='o', s=100)
axes[1, 0].set_title("Scatter Plot")

# Fourth subplot: Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True)
axes[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
